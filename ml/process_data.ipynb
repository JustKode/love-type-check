```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
df = pd.read_csv('./data.csv')
df
```

```python
string_to_value = {
    "매우 아니다.": -1.0,
    "아니다.": -0.5,
    "보통이다.": 0.0,
    "그렇다.": 0.5,
    "매우 그렇다.": 1.0
}

data = df.iloc[:,1:21].applymap(lambda x: string_to_value[x]).to_numpy()
target = df.iloc[:, -1].map(lambda x: np.array(list(map(float,(x.split(','))))))
data, target = torch.Tensor(data), torch.Tensor(target)
```

```python
class LinearModel(nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.linear = nn.Linear(20, 4)
        self.softmax = nn.Softmax()
    
    def forward(self, x):
        x = self.linear(x)
        x = self.softmax(x)
        return x
```

```python
model = LinearModel()

criterion = nn.MSELoss(reduction='mean')
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
for epoch in range(100):
    y_pred = model(data)
    loss = criterion(y_pred, target)
    print(f"{epoch + 1} Epoch | Loss: {loss.item()}")
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
def type_result(result):
    idx_to_type = {0: "공생", 1: "쾌락", 2: "생산", 3: "소유"}
    _, idx = result.sort(descending=True)
    return f"{idx_to_type[int(idx[0])]}-{idx_to_type[int(idx[1])]}"

new_dataframe = df.iloc[:, 21: 26]
new_dataframe['result'] = pd.Series([type_result(r) for r in model(data)])
new_dataframe.to_excel('./result.xlsx')
new_dataframe
```
